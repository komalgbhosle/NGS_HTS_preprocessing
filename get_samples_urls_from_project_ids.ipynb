```python
import pandas as pd
import time
import re
import requests
import pprint 
import json
import csv
from bs4 import BeautifulSoup
from nested_lookup import nested_lookup
from tqdm import tqdm
from operator import itemgetter 
import urllib.request, json
import urllib, json
```

```python
df = pd.read_csv("bioproject_study_ids.csv", header = None)
df
```

```python
study_ids = df.values.tolist()
study_ids
```

```python
Study_id = []
for sublist in study_ids:
    for item in sublist:
        Study_id.append(item)
```

```python
Study_ids = set(Study_id)
#Study_ids = ['PRJNA642155','PRJNA271784']
```

```python
newfastqDictionary_Seq = {}
rna_seq_samples_urls = []
for ID in Study_ids:
        try:
            fastq_file_json = 'https://www.ebi.ac.uk/ena/portal/api/filereport?accession={}&result=read_run&fields=study_accession,sample_accession,experiment_accession,run_accession,tax_id,scientific_name,fastq_ftp,submitted_ftp,sra_ftp&format=json'.format(ID)
            page = requests.get(fastq_file_json)
            newfastqDictionary_Seq = json.loads(page.text)
            #print(newfastqDictionary_Seq)
            for i in newfastqDictionary_Seq:
                urls = i['fastq_ftp']
                rna_seq_samples_urls.append((ID,urls))
        except ValueError:
            rna_seq_samples_urls.append((ID,"content not found"))
print(rna_seq_samples_urls)
```

```python
fields = ['Project_ids', 'urls']
with open('indication_sample_urls', 'w') as f:
    write = csv.writer(f) 
    write.writerow(fields) 
    write.writerows(rna_seq_samples_urls) 
```

```python
df = pd.DataFrame(rna_seq_samples_urls)
```

```python
df.to_csv("indication_sample_urls.csv", index = False)
```
